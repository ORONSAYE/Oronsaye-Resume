```python
%%writefile run_gprmax_gpu.py

import os
import subprocess
import shutil

# Define the project directory (replace with your unzipped folder path in Colab)
project_dir = "/content/NewProj"  # Update this if your project directory name is different
input_dir = os.path.join(project_dir, "gprmax_inputs")  # Directory containing input files
gprmax_outputs_dir = os.path.join(project_dir, "gprmax_outputs")  # Folder to save merged output files
log_file = os.path.join(project_dir, "last_processed.txt")  # File to store the last processed input file

# Create gprmax_outputs directory if it doesn't exist
os.makedirs(gprmax_outputs_dir, exist_ok=True)

# Get a list of all input files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith(".in")]

# Check if we have a log file that stores the last processed input file
last_processed = ""
if os.path.exists(log_file):
    with open(log_file, 'r') as file:
        last_processed = file.read().strip()

# If no file is recorded or log file is empty, start from the first file
if last_processed == "":
    start_processing = True
else:
    start_processing = False

# Loop over all input files and run the GPRMax simulation for each
for input_file in input_files:
    # Skip files until we reach the next unprocessed file
    if not start_processing:
        if input_file == last_processed:
            start_processing = True
        else:
            continue

    input_path = os.path.join(input_dir, input_file)

    # Prepare the command to run the GPRMax simulation with GPU support
    command = [
        "python3", "-m", "gprMax", input_path, "-n", "476", "-gpu"
    ]

    # Run the simulation
    print(f"Running GPU simulation for {input_file}...")
    subprocess.run(command, check=True)

    # Remove the file extension (.in) from input file name to use as a base name
    base_name = os.path.splitext(input_file)[0]

    # Prepare the command to merge the output files
    merge_command = [
        "python3", "-m", "tools.outputfiles_merge", os.path.join(input_dir, base_name), "--remove-files"
    ]

    # Run the merging command
    print(f"Merging output files for {base_name}...")
    subprocess.run(merge_command, check=True)

    # Immediately move the merged output file to the 'gprmax_outputs' folder
    merged_file_path = os.path.join(input_dir, f"{base_name}_merged.out")

    if os.path.exists(merged_file_path):
        # Move (cut) the merged file directly to the 'gprmax_outputs' directory
        new_merged_file_path = os.path.join(gprmax_outputs_dir, f"{base_name}_merged.out")

        try:
            shutil.move(merged_file_path, new_merged_file_path)
            print(f"Moved merged file for {base_name} to the 'gprmax_outputs' folder.")
        except Exception as e:
            print(f"Error moving merged file for {base_name}: {e}")
    else:
        print(f"Warning: Merged file for {base_name} not found.")

    # Clean up intermediate files after merging
    intermediate_dir = os.path.join(input_dir, base_name)
    if os.path.exists(intermediate_dir):
        # Delete all intermediate files in the base_name directory
        for file in os.listdir(intermediate_dir):
            file_path = os.path.join(intermediate_dir, file)
            try:
                os.remove(file_path)
                print(f"Deleted intermediate file: {file}")
            except Exception as e:
                print(f"Error deleting file {file}: {e}")

        # Remove the directory itself after cleaning up intermediate files
        try:
            os.rmdir(intermediate_dir)
            print(f"Deleted intermediate directory: {intermediate_dir}")
        except Exception as e:
            print(f"Error deleting directory {intermediate_dir}: {e}")

    # Log the last processed file to resume next time
    with open(log_file, 'w') as file:
        file.write(input_file)

print(f"All GPU simulations and file merging completed. Merged files are in the '{gprmax_outputs_dir}' directory.")
```
